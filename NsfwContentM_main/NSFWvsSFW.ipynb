```python
import re
import unicodedata
import emoji
import html
from ftfy import fix_text

# Compile regex patterns once (performance improvement)
URL_PATTERN = re.compile(r'https?://\S+|www\.\S+')
RT_PATTERN = re.compile(r'[\W_]*rt\s+@\w+:?', flags=re.IGNORECASE)
MENTION_PATTERN = re.compile(r'@\w+')
ELONG_PATTERN = re.compile(r'(.)\1{2,}')
EXCL_PATTERN = re.compile(r'!{3,}')
QUES_PATTERN = re.compile(r'\?{3,}')
MULTISPACE_PATTERN = re.compile(r'\s+')

# Emoji semantic buckets (substring matching)
EMO_THREAT = re.compile(
    r':[^:]*(gun|pistol|knife|dagger|bomb|skull|coffin|blood|fire|explod|sword)[^:]*:'
)

EMO_NEG = re.compile(
    r':[^:]*(angry|rage|mad|enrag|symbol|horn|imp|ogre|goblin|poo|vomit|nauseat|clown)[^:]*:'
)

EMO_POS = re.compile(
    r':[^:]*(smile|grin|joy|laugh|heart|thumb|clap|love|wink)[^:]*:'
)

EMO_OTHER = re.compile(r':[a-z0-9_]+:')


def split_hashtag(match):
    tag = match.group()[1:]
    tag = re.sub(r'([a-z])([A-Z])', r'\1 \2', tag)
    return " " + tag.lower()


def preprocess_text(text):
    text = str(text)

    # Fix encoding issues
    text = fix_text(text)

    # Unicode normalization
    text = unicodedata.normalize("NFKC", text).strip()

    # Decode HTML entities
    text = html.unescape(text)

    # Lowercase
    text = text.lower()

    # Remove URLs
    text = URL_PATTERN.sub(' ', text)

    # Remove RT metadata
    text = RT_PATTERN.sub(' ', text)

    # Remove mentions
    text = MENTION_PATTERN.sub(' ', text)

    # Process hashtags
    text = re.sub(r'#\w+', split_hashtag, text)

    # Normalize elongations (soooo → soo)
    text = ELONG_PATTERN.sub(r'\1\1', text)

    # Normalize punctuation intensity
    text = EXCL_PATTERN.sub('!!', text)
    text = QUES_PATTERN.sub('??', text)

    # ---- Emoji handling ----
    text = emoji.demojize(text, language='en')

    text = EMO_THREAT.sub(' <EMO_THREAT> ', text)
    text = EMO_NEG.sub(' <EMO_NEG> ', text)
    text = EMO_POS.sub(' <EMO_POS> ', text)

    # Replace remaining emojis
    text = EMO_OTHER.sub(' <EMO_OTHER> ', text)

    # Final cleanup
    text = MULTISPACE_PATTERN.sub(' ', text).strip()

    return text


# Apply preprocessing
df['text_clean'] = df['text'].apply(preprocess_text)

```

```python
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split

df = shuffle(df, random_state=42).reset_index(drop=True)


train_df, temp_df = train_test_split(
    df,
    test_size=0.30,
    stratify=df['label'],
    random_state=42
)

val_df, test_df = train_test_split(
    temp_df,
    test_size=0.50,
    stratify=temp_df['label'],
    random_state=42
)

def check_distribution(df, name):
    print(f"\n{name} distribution:")
    print(df['label'].value_counts(normalize=True))

check_distribution(train_df, "Train")
check_distribution(val_df, "Validation")
check_distribution(test_df, "Test")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

MAX_VOCAB = 30000
MAX_LEN   = 50
EMB_DIM   = 128
LSTM_UNITS = 64
NUM_CLASSES = 3

tokenizer = Tokenizer(
    num_words=30000,
    oov_token="<OOV>",
    filters='!"#$%&()*+,-./:;=?@[\\]^_`{|}~\t\n',  # REMOVED < >
    lower=True
)

# Fit ONLY on training data (NO LEAKAGE)
tokenizer.fit_on_texts(train_df["text_clean"])

# Convert text to sequences
X_train_seq = pad_sequences(
    tokenizer.texts_to_sequences(train_df["text_clean"]),
    maxlen=MAX_LEN,
    padding="post",
    truncating="post"
)

X_val_seq = pad_sequences(
    tokenizer.texts_to_sequences(val_df["text_clean"]),
    maxlen=MAX_LEN,
    padding="post"
)

X_test_seq = pad_sequences(
    tokenizer.texts_to_sequences(test_df["text_clean"]),
    maxlen=MAX_LEN,
    padding="post"
)

y_train = train_df["label"].astype(int).values
y_val   = val_df["label"].astype(int).values
y_test  = test_df["label"].astype(int).values
```

```python
import numpy as np
from sklearn.utils.class_weight import compute_class_weight

classes = np.unique(y_train)
weights = compute_class_weight(
    class_weight="balanced",
    classes=classes,
    y=y_train
)

class_weights = dict(zip(classes, weights))
class_weights
```

```python
from tensorflow.keras.layers import Layer
class SelfAttention(Layer):
    def build(self, input_shape):
        self.context = self.add_weight(
            name="context_vector",
            shape=(input_shape[-1],),
            initializer="glorot_uniform",
            trainable=True
        )

    def call(self, inputs):
        scores = tf.tensordot(inputs, self.context, axes=1)   # (batch, time)
        weights = tf.nn.softmax(scores, axis=1)               # (batch, time)
        output = tf.reduce_sum(inputs * tf.expand_dims(weights, -1), axis=1)
        return output
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, Embedding, Bidirectional, LSTM,
    Dense, Dropout, BatchNormalization,
    SpatialDropout1D, Attention, GlobalAveragePooling1D
)
from tensorflow.keras import regularizers

inputs = Input(shape=(MAX_LEN,))

x = Embedding(
    input_dim=MAX_VOCAB,
    output_dim=EMB_DIM,
    mask_zero=True
)(inputs)

x = Bidirectional(
    LSTM(LSTM_UNITS, return_sequences=True)
)(x)

x = SelfAttention()(x)
x = Dropout(0.5)(x)

outputs = Dense(NUM_CLASSES, activation="softmax")(x)

model = Model(inputs, outputs)
```

```python
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
early_stop = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    patience=3,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.5,
    patience=2,
    min_lr=1e-6,
    verbose=1
)
callbacks = [early_stop, reduce_lr]
```

```python
history = model.fit(
    X_train_seq, y_train,
    validation_data=(X_val_seq, y_val),
    epochs=20,
    batch_size=64,
    callbacks=callbacks,
    class_weight=class_weights
)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = np.argmax(model.predict(X_test_seq), axis=1)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred, digits=4))
```

```python
import numpy as np
from tensorflow.keras.preprocessing.sequence import pad_sequences

# ===============================
# Label mapping (YOUR FORMAT)
# ===============================
label_map = {
    0: "neutral",
    1: "offensive",
    2: "hate_speech"
}

# ===============================
# Test sentences + actual labels
# ===============================
test_data = [
    # =========================
    # HATE SPEECH (2)
    # Group-targeted, dehumanizing
    # =========================
    ("Those people are parasites destroying society", 2),
    ("That group deserves to be wiped out completely", 2),
    ("People like them are subhuman garbage", 2),
    ("That community should be banned from existing", 2),
    ("They are a plague and must be eliminated", 2),

    # =========================
    # OFFENSIVE (1)
    # Personal attacks, abuse
    # =========================
    ("You are a worthless piece of trash", 1),
    ("Your brain is empty and you are useless", 1),
    ("This post is absolute garbage, idiot", 1),
    ("You are painfully stupid and annoying", 1),
    ("Shut up, nobody wants to hear your nonsense", 1),
    ("fuck you bitch",1),

    # =========================
    # NEUTRAL (0)
    # Safe, normal content
    # =========================
    ("I am training an LSTM for text classification", 0),
    ("This application lets users upload images", 0),
    ("Machine learning models require good data", 0),
    ("I am debugging prediction errors", 0),
    ("Can someone explain how tokenization works", 0),
    ("we will go for dinner tonight",0)
]

# ===============================
# Same max length as training
# ===============================
# MAX_LEN = 50

# ===============================
# Inference
# ===============================
for sentence, actual_label in test_data:
    seq = tokenizer.texts_to_sequences([sentence])
    padded = pad_sequences(seq, maxlen=MAX_LEN, padding="post")

    pred = model.predict(padded, verbose=0)
    pred_label = np.argmax(pred, axis=1)[0]

    print("Sentence :", sentence)
    print("Actual   :", label_map[actual_label])
    print("Predicted:", label_map[pred_label])
    print("-" * 50)
```
